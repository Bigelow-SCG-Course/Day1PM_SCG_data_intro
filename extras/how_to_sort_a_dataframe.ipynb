This notebook shows you how to sort a dataframe and examine it in a Jupyter notebook.

```python
import pandas as pd

# path to your csv here:

my_file = "/mnt/storage/data/AG-910_assembly_stats_updated.csv"

# load your file into a dataframe:
df = pd.read_csv(my_file)

# look at the columns of your table
df.columns
```

```python
# sort your dataframe

# this should be one of your column names.  Replace 'final_assembly_length' with whatever value you want to sort by:
sort_value = 'final_assembly_length'

# do you want the sort to be ascending?  If you want it to be ascending (low to high), keep as is.  If not, change this to False
ascending = True

# How many rows do you want to look at after you've sorted your dataframe?
rowcount = 20

# sort command:
df.sort_values(by = sort_value, ascending = ascending).head(rowcount)
```
